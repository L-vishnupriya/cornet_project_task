# **WER**

```python
from jiwer import wer

correct = "Take two tablets daily"
Predicted = "Take tablet daily"

error = wer(correct, Predicted)
print("WER:", error)
```

**Torch**

```python
import torch
from torchmetrics.text import WordErrorRate

wer_metric = WordErrorRate()

reference = ["Take two tablets daily"]
hypothesis = ["Take tablet daily"]

error = wer_metric(hypothesis, reference)
print("WER:", error.item())
```

**Sentence Error Rate(SER):**

```python
reference = [
    "The patient is stable",
    "Take two tablets daily",
    "No history of diabetes"
]

hypothesis = [
    "The patient is stable",
    "Take two tablet daily",
    "No history of diabetes"
]

error_sentences = 0

for r, h in zip(reference, hypothesis):
    if r != h:
        error_sentences += 1

ser = error_sentences / len(reference)
print("SER:", ser)
```
